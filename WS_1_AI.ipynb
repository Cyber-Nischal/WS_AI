• For the following arrays:
A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);
Prove following with Numpy:
1. Prove A.A−1 = I.

```python
import numpy as np

A = np.array([[3,4],[7,8]])

A_inv = np.linalg.inv(A)

result = np.dot(A, A_inv)

result_rounded = np.round(result)

print("A * A^-1 =\n", result_rounded)
```

2. Prove AB ̸= BA.

```python
import numpy as np

A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

AB = np.dot(A, B)
BA = np.dot(B, A)

print("A * B =\n", AB)
print("\nB * A =\n", BA)
print("\nCheck if AB equals BA:", np.array_equal(AB, BA))
```

3. Prove (AB)

T = BTAT

```python
import numpy as np

A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

# Compute (AB)^T
AB_T = np.dot(A, B).T

# Compute B^T * A^T
B_TA_T = np.dot(B.T, A.T)

print("(AB)^T =\n", AB_T)
print("\nB^T * A^T =\n", B_TA_T)
print("\nCheck if (AB)^T equals B^T*A^T:", np.array_equal(AB_T, B_TA_T))
```

• Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9

{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}

```python
import numpy as np

A = np.array([
    [ 2, -3,  1],
    [ 1, -1,  2],
    [ 3,  1, -1]
], dtype=float)

B = np.array([-1, -3, 9], dtype=float)

A_inv = np.linalg.inv(A)
X = A_inv.dot(B)

print(X)  # [x, y, z]
```

• Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” function
of Numpy}

```python
import numpy as np

# Coefficient matrix A
A = np.array([
    [ 2, -3,  1],
    [ 1, -1,  2],
    [ 3,  1, -1]
], dtype=float)


B = np.array([-1, -3, 9], dtype=float)

A_inv = np.linalg.inv(A)

X = np.dot(A_inv, B)

print("Solution using np.linalg.inv():")
print("x =", X[0])
print("y =", X[1])
print("z =", X[2])
```
